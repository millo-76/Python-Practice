```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("http://128.198.162.191/dataFiles/HURRICANES/hurdat2-1851-2023-051124.txt")

df.head()
```

### Add colum names

```python
filepath = "http://128.198.162.191/dataFiles/HURRICANES/hurdat2-1851-2023-051124.txt"

column_names = [
    "ID-Date", "Time", "Record Identifier", "Status", "Latitude", "Longitude",
    "Windspeed", "Pressure", "extra1", "extra2", "extra3", "extra4", "extra5",
    "extra6", "extra7", "extra8", "extra9", "extra10", "extra11", "extra12"
]

df = pd.read_csv(filepath, header=0, names=column_names, usecols=range(len(column_names)), on_bad_lines='skip')

df.head()
```

### Extract "ID", "YEAR", and "MONTH"

```python
# Create colums to cteate "ID", "YEAR", and "MONTH"
df['YEAR'] = ""
df['MONTH'] = ""

# only extract year and month where ID-Date doesn not contain AL
df.loc[~df['ID-Date'].str.contains('AL'), 'YEAR'] = df['ID-Date'].str[:4]
df.loc[~df['ID-Date'].str.contains('AL'), 'MONTH'] = df['ID-Date'].str[4:6]

# create a new  UNIQUE_ID column
df['UNIQUE_ID'] = df['ID-Date'].apply(lambda x: x if 'AL' in x else None)

# forward fill the UNIQUE_ID column to propagate the hurricane ID to all rows
df['UNIQUE_ID'] = df['UNIQUE_ID'].ffill()

df.head()
```

### Total hurricanes per year

```python
hurricanesByYear = df.groupby('YEAR')['UNIQUE_ID'].count().reset_index(name='COUNT')

hurricanesByYear
```

### Visualize by Year

```python
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')

# drop rows with NaN values in 'YEAR' column
dfClean = df.dropna(subset=['YEAR'])

# convert 'YEAR' column to integer type
dfClean.loc[:, 'YEAR'] = dfClean['YEAR'].astype(int)

# group the cleaned DataFrame
hurricanesByYear = dfClean.groupby('YEAR')['UNIQUE_ID'].nunique()

# plot results
plt.figure(figsize=(15, 10))
plt.plot(hurricanesByYear.index, hurricanesByYear.values, marker='o')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Total Number of Hurricanes by Year')

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```

```python
# add averasge line

# plot results
plt.figure(figsize=(15, 5))
plt.plot(hurricanesByYear.index, hurricanesByYear.values, marker='o')

# add an average line
cumulative_avg = hurricanesByYear.expanding().mean()

plt.plot(cumulative_avg.index, cumulative_avg.values, color='red', linestyle='--', label='Cumulative Average')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Total Number of Hurricanes by Year')

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```
