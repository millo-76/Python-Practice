## research questions

### search for passenger

```python
# search for molly brown

molly_brown = df[df['Name'].str.contains('Brown', case=False)]

print(molly_brown)
```

### Who survived?

```python
# group by 'Survived' and name

who_survived = df.groupby('Survived')['Name'].apply(list)

print("List of survivors.", who_survived[1])
print("List of those who did not survive.", who_survived[0])
```

```python
who_survived = df.groupby('Survived')['Name'].apply(list)

print("List of survivors: ")

for name in who_survived[1]:
    print(name)

print("\nList of those who did not survive: ")

for name in who_survived[0]:
    print(name)
```

### how many men vs women survived

```python
survivedByGender = df.groupby(['Survived', 'Sex']).size()

print(survivedByGender)
```

### Odd of surviving men vs women

```python
maleSurvived = df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0]
maleDied = df[(df['Sex'] == 'male') & (df['Survived'] == 0)].shape[0]

femaleSurvived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)].shape[0]
femaleDied = df[(df['Sex'] == 'female') & (df['Survived'] == 0)].shape[0]

oddsMale = maleSurvived / maleDied

oddsFemale = femaleSurvived / femaleDied

print(f"Odds of surviving for men: {oddsMale:.2f}")
print(f"Odds of surviving for women: {oddsFemale:.2f}")
```

## Resqrch Questions

#### Name and year of movie with highest rating and highest number of votes.

```python
# find the data; dfratings for votes and averageRating, dftitles for primaryTitle and startYear

dfmerged = pd.merge(dftitles, dfratings, on='tconst')   # merges based on tconst column common to both

dfmerged
```

```python
rating_sort = (dfmerged.query('titleType == "movie"'))

rating_sort
```

```python
rating_sort.sort_values(by= 'averageRating', ascending=False)
```

```python
rating_sort.sort_values(by= ['averageRating', 'numVotes'], ascending=False)
```

```python
# reduce information to only what is needed

semi1 = (rating_sort.sort_values(by=['numVotes', 'averageRating'], ascending=False))

# select only relevant columns

semi2 = semi1[['primaryTitle', 'startYear', 'numVotes', 'averageRating']]

# select a subset of rows

final = (semi2.head())

print(final)
```
